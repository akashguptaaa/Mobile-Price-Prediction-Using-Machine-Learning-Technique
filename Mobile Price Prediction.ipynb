```python
#Step 3: here i have been dividing the data into two variable x and y

X = dataset.iloc[:,:-1].values

y = dataset.iloc[:,1].values

#X = dataset.drop('price_range',axis=1)

#y = dataset['price_range']


```

```python
X.shape
```

```python
y.shape
```

```python
# here i am splitting the data into training and testing. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10,random_state=10)
```

```python
# here i am implementing the Linear Regression model.
from sklearn.linear_model import LinearRegression
linearRegressor =   LinearRegression() 
linearRegressor.fit(X_train,y_train)
y_pred = linearRegressor.predict(X_test)
```

```python
#R-square score
linearRegressor.score(X_test,y_test)
```

```python
y_pred
```

```python
# i am using Root mean square error to check the varience between actual and prediction.
from sklearn import metrics
print('RMSE',np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
